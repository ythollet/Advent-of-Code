# Import des librairies

```python
import numpy as np
import re
from collections import namedtuple
```

# Lecture du fichier source + nettoyage

```python
data = []
tuple_instruction = namedtuple(
    'tuple_instruction',
    ['method','l1','r1','l2','r2']
)   

with open('real.txt') as fichier:
    for line in fichier:
        
        # Nettoyage
        ligne = line.strip().replace('through ','').replace('turn ','')
        ligne = re.findall(r'\w+|\d+', ligne)

        # Déballage
        method,l1,r1,l2,r2 = ligne
        current_tuple = tuple_instruction(method, int(l1), int(r1), int(l2), int(r2))
        data.append(current_tuple)
data
```

# Traitement

```python
dim_file = 1000
matrix = np.zeros((dim_file, dim_file), dtype=int)
matrix

# Pour chaque instruction (sous forme de NamedTuple)
for instruction in data:

    # Zone a modfier
    area = matrix[instruction.l1 : instruction.l2 +1, 
               instruction.r1 : instruction.r2 +1]
    
    # Augmenter luminosté
    if instruction.method == 'on':
        area += 1 

    # Diminuer luminosité
    elif instruction.method == 'off':
        area -= 1
        np.clip(area, 0, None, out=area)
        
    # Augmenter de 2 la luminosté
    elif instruction.method == 'toggle':
        area += 2
        
print(matrix.sum())
        


```
