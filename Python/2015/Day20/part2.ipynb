# Import des librairies

```python
from collections import defaultdict
import math
```

```python
def func_get_firsts_50_multiples(
    in_number: int
) -> set:       
    
    return {in_number*m for m in range(1,50)}
```

```python
def func_in_50_firsts_multiples(
    in_number: int,
    in_target: int
) -> bool:
    
    return in_target in func_get_firsts_50_multiples(in_number=in_number)
```

```python
def func_get_diviseurs(
    number:int
) -> list[int]:
    
    diviseurs = set()
    
    for div in range(1,int(math.sqrt(number))+1):
        if (number % div) == 0:
            other_div = number // div

            # Un diviseur est ajouté uniquement si le numéro de la maison est compris dans ses 50 premiers multiples
            if func_in_50_firsts_multiples(in_number=number,in_target=div):
                diviseurs.add(div)
            
            if func_in_50_firsts_multiples(in_number=number,in_target=other_div):
                diviseurs.add(other_div)
    
    return list(diviseurs)
```

```python
func_get_diviseurs(2)
```

```python
def func_get_nb_presents_for_a_house(
    in_num_house: int
) -> int :

    list_num_elfs = func_get_diviseurs(in_num_house)

    nb_presents_house = sum(11*elf_number for elf_number in list_num_elfs)


    return nb_presents_house
```

```python
target_nb_presents = 29000000
num_house = 1
nb_presents_house = 0 

while True:
    
    nb_presents_house = func_get_nb_presents_for_a_house(num_house)

    if nb_presents_house >= target_nb_presents:
        break
    num_house += 1

num_house
```
