# Lecture du fichier source

```python
with open('real.txt') as fichier:
    file = fichier.read()
file
```

# Traitement

```python
pos = (0,0)
visited = {(0,0)}

dict_moove = {
    '>' : (0,1),
    'v' : (1,0),
    '<' : (0,-1),
    '^' : (-1,0)
}

# Pour chaque instruction 
for char in file:
    
    # On récupère le déplacement à effectuer
    delta = dict_moove[char]

    # On met à jour de la position
    pos = (pos[0] + delta[0], pos[1] + delta[1])

    # On ajoute la position à l'ensemble des positions visitées
    visited.add(pos)
```
