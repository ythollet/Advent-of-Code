# Import des librairies

```python
import hashlib
```

# Lecture du fichier source

```python
with open('real.txt') as fichier:
    file = fichier.read()
file
```

```python
number = 1

while True:
    # Concaténation de la clé secrète avec le nombre
    message = file + str(number)

    # Calcul du hash MD5
    hash_result = hashlib.md5(message.encode()).hexdigest()

    # Vérifie si le hash commence par cinq zéros
    if hash_result.startswith('00000'):
        print(number)
        break

    number += 1
```
