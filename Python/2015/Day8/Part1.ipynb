# Import des librairies

```python
import re
```

# Lecture du fichier source

```python
with open('real.txt') as fichier:
    file = fichier.read().split('\n')
file
```

```python
total = 0

for row in file:
    print((row))
    print('longueur row',len(row))
    longueur_brute = len(row) # 2 car on compte les guillemets

    # Supprimer les "" du debut et de fin
    row = row[1:-1]
    print('suppression des gillemets de debut / fin : ',row)

    # Détecter les \x 
    list_code_ascii = re.findall(r'\\x[0-9a-f]{2}', row)


    # Supprimer les element de la chaine contenus dans list_code_ascii
    for elem_to_supp in list_code_ascii:
        row = row.replace(elem_to_supp,'A')
    print('suppression des backslahsh x : ',row)

    # Supprime les \\
    row = row.replace('\\\\','B')
    print('suppression des double backslash : ',row)

    # Supprimer les \"
    row = row.replace('\\"','C')
    print('suppression des double backslash duouble quote : ',row)

    print('longueur row finale',len(row))

    total += longueur_brute - len(row)
    print()

total
```
