# Lecture du fichier source

```python
with open('real.txt') as fichier:
    file = fichier.read()
```

# Traitement

```python
num_instructions = len(file)
num_open_parenthesis = sum(1 for char in file if char=='(')
num_closed_parenthesis = num_instructions - num_open_parenthesis

num_etage = num_open_parenthesis - num_closed_parenthesis
num_etage
```
