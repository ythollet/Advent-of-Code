# Lecture du fichier source

```python
with open("real.txt") as fichier:
    file = fichier.read()
    data = json.loads(file)
data
```

```python
def func_explore(
    in_data: list | dict
) -> int:
    """
    Parcourt de manière récursivé les structures imbriquées (listes, dictionnaires)

    Args:
        in_data (list | dict): structure à parcourir

    Returns:
        int: somme de tous les nombres du fichier source
    """

    somme = 0
    data_is_dict = False

    # Si l'itérable est un Dict
    if isinstance(in_data, dict):
        data_is_dict = True

        # On le convertis en Liste
        in_data = list(in_data.values())

    # Parcours tous les elements de la Liste 
    for element in in_data:

        # Si de base on avait un Dict, mais qu'on a convertit en List
        if data_is_dict:
            
            # On exclus le dict si il contient 'red'
            if element == 'red':
                return 0

        # Si l'élément est un String
        if isinstance(element, str):
            continue
    
        # Si l'élément est un Int
        elif isinstance(element, int):

            # On ajoute sa valeur à une variable
            somme += element

        # Si l'élément est un Dict ou une Liste
        else:
            somme += func_explore(
                in_data = element
            )

    return somme
```

```python
func_explore(
    in_data = data,
)
```
