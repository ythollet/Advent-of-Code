# Préparation au traitement

```python
# On définit l'alphabet
alphabet = "abcdefghijklmnopqrstuvwxyz"

# On stocke dans un dict:
# - les clés représente chaque lettre
# - les valeurs representent la lettre suivante de cette lettre
dict_next_letter = {
    alphabet[pos_letter]: alphabet[(pos_letter+1)%26] 
    for pos_letter 
    in range(len(alphabet))
    }

dict_next_letter
```

```python
def func_increment_password(
    in_password: list[str],
    in_dict_next_letter: dict
) -> list[str]:
    """
    Incrémente le mot de passe:
    exemple d'incrémentations:  xx -> xy -> xz -> ya -> yb...

    Args:
        in_password (list[str]): mot de passe
        in_dict_next_letter (dict): dictionnaire qui, pour une lettre donnée, donne la lettre suivante

    Returns:
        str: mot de passe incrémenté
    """
    new_password = in_password.copy()
 
    # Si la dernière lettre du mot de passe est 'z', il faut potentiellement gérer des incrémentations en cascade
    if in_password[-1] == 'z':
        
        # On parcours le mot de passe à l'envers
        for pos_letter, letter in zip(range(len(in_password)-1,-1,-1),reversed(in_password)):

            # Si la lettre est 'z'
            if letter == 'z':

                # On incrémente la lettre positionnée juste avant
                new_password[pos_letter-1] = in_dict_next_letter[new_password[pos_letter-1]]
            
            # Pour la premère lettre différente de 'z' trouvée (en parcourant le mdp à l'envers)
            else:
                # On arrete la boucle for 
                break

    # On incrémente la dernière lettre du mot de passe
    new_password[-1] = in_dict_next_letter[in_password[-1]]

    return new_password
```

```python
def check_password_conditions(
    in_password: list[str]
) -> bool:
    
    # 1ere condition
    first_condition = any(
        ord(l2) == ord(l1)+1 and ord(l3) == ord(l2)+1
        for l1, l2, l3
        in zip(in_password, in_password[1:], in_password[2:])
    )

    # Test de la 1ere condition
    if first_condition is False:
        # print('le mot de passe ne contient pas de suite (ex:abc)')
        return False


    # 2e condition
    forbidden_letters = {'i','o','l'}
    second_condition = not any(
        forbidden_letter in in_password
        for forbidden_letter
        in forbidden_letters
    )
    
    # Test de la 2e condition
    if second_condition is False:
        # print("le mot de passe contient une lettre interdite: 'i','o','l'")
        return False
    
    # 3e condition
    doubles = {
        l1
        for l1 ,l2
        in zip(in_password, in_password[1:])
        if l1 == l2
    }
    third_condition = len(doubles) >= 2
  
    # Test de la 3e condition
    if third_condition is False:
        # print("le mot de passe ne contient pas deux doublons")
        return False
    

    return True
```

```python
continue_loop = True
nb_password = 0

while continue_loop:

    new_mdp = func_increment_password(
        in_password = password,
        in_dict_next_letter = dict_next_letter
    )
    if check_password_conditions(new_mdp):
        nb_password += 1
        if nb_password == 2:
            continue_loop = False

    password = new_mdp.copy()

''.join(new_mdp)
```
