# Import des librairies

```python
from dataclasses import dataclass
import re
```

# Lecture du fichier source

```python
with open('real.txt') as data:
    file = data.read().split("\n")
file
```

# Classe

```python
@dataclass
class Reindeer:
    vmax_km_s: int
    time_vmax: int
    time_recup: int
```

```python
def func_transform_row(
    in_row: str
) -> Reindeer:
    
    # Récupère les infos de la ligne
    vmax_km_s, time_vmax, time_recup = re.findall(r"\d+", in_row)

    # Conversion en objet Reindeer
    obj_reindeer =  Reindeer(
        vmax_km_s = int(vmax_km_s),
        time_recup = int(time_recup),
        time_vmax = int(time_vmax)
    )

    return obj_reindeer
```

```python
list_obj_reindeer = [func_transform_row(row) for row in file]
list_obj_reindeer
```

```python
def func_calcul_distance(
    in_obj_reindeer: Reindeer,
    in_time: int
) -> int:
    """
    Calcule la distance parcourue par un renne apres un certain temps
    """

    remaining_time = in_time
    total_distance = 0

    while remaining_time > 0:
        print("temps restant : ",remaining_time)
        print("distance : ", total_distance)

        # Si le renne n'a pas le temps de faire un sprint entier
        if remaining_time <= in_obj_reindeer.time_vmax:
            print("le renne n'a pas le temps de faire un sprint entier")
            total_distance += in_obj_reindeer.vmax_km_s * remaining_time
            break

        # Si le renne a le temps de faire un sprint entier
        else:
            print("le renne a le temps de faire un sprint entier")
            remaining_time -= in_obj_reindeer.time_vmax
            print("")
            total_distance += in_obj_reindeer.vmax_km_s * in_obj_reindeer.time_vmax
            remaining_time -= in_obj_reindeer.time_recup
        print('distance : ',total_distance)
        print()


    return total_distance
```
