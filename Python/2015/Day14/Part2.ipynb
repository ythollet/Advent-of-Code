# Import des librairies

```python
from dataclasses import dataclass
import re
```

# Lecture du fichier source

```python
with open('real.txt') as data:
    file = data.read().split("\n")
file
```

# Classe

```python
@dataclass
class Reindeer:
    name: str
    vmax_km_s: int
    time_vmax: int
    time_recup: int
    time_cycle: int
    current_distance: int = 0
    current_score: int = 0
```

```python
def func_transform_row(
    in_row: str
) -> Reindeer:
    
    # Récupère les infos de la ligne
    vmax_km_s, time_vmax, time_recup = re.findall(r"\d+", in_row)

    # Conversion en objet Reindeer
    obj_reindeer =  Reindeer(
        name = in_row.split()[0],
        vmax_km_s = int(vmax_km_s),
        time_recup = int(time_recup),
        time_vmax = int(time_vmax),
        current_distance = 0,
        time_cycle = int(time_recup) + int(time_vmax)
    )

    print(obj_reindeer.time_cycle)


    return obj_reindeer
```

```python
list_obj_reindeer = [func_transform_row(row) for row in file]
list_obj_reindeer
```

```python
def func_calcul_best_score(
    in_list_obj_reindeer: list[Reindeer],
    in_time: int
) -> int:
    """
    """

    # Pour chaque seconde qui s'écoule
    for current_time in range(1,in_time+1):
        print("current_time")
        print(current_time)
        
        # Pour chaque renne de la liste
        for reindeer in in_list_obj_reindeer:
            
            # Calcule le temps modulo le temps d'un cycle 
            time_modulo_cycle_reindeer = current_time % reindeer.time_cycle
            
            # Si le Renne est dans sa période de sprint
            if 0 < time_modulo_cycle_reindeer <= reindeer.time_vmax:

                # Il avance
                reindeer.current_distance += reindeer.vmax_km_s

            # Si le Renne est dans sa période de repos
            else:

                # Il ne bouge pas
                pass



        # On détermine la distance parcourue par le/les leaders
        distance_leaders = max(
            reindeer.current_distance
            for reindeer
            in list_obj_reindeer
        )

        # Pour chaque renne
        for reindeer in list_obj_reindeer:

            # Si le rennne est leader
            if reindeer.current_distance == distance_leaders:

                # On incrémente son score
                reindeer.current_score += 1
            print(reindeer)
        print()


    return max(reindeer.current_score for reindeer in in_list_obj_reindeer)
```
