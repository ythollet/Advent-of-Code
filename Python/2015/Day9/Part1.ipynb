# Import des librairies

```python
import re
from dataclasses import dataclass
from itertools import permutations
import time
```

# Classe Trajet

```python
@dataclass
class Trajet:
    ville1: str
    ville2: str
    distance: int
```

# Fonction nettoyage ligne

```python
def func_line_to_obj_trajet(
    in_line: str
    ) -> Trajet:
    """ 
    Transforme une ligne du fichier source en objet Trajet

    Args:
        in_line (str): ligne à transformer

    Returns:
        Trajet: objet Trajet
    """

    parts = re.split(r'to|=', in_line)
    
    obj_trajet = Trajet(
        ville1 = parts[0].strip(),
        ville2 = parts[1].strip(),
        distance = int(parts[2].strip())
    )

    return obj_trajet
```

# Lecture du fichier source

```python
with open('real.txt') as fichier:
    
    liste_trajets = [
        func_line_to_obj_trajet(line)
        for line 
        in fichier.read().split('\n')
        ]
liste_trajets
```

# Ajout de la version inversée du trajet

```python
liste_trajets_inverse = [
    Trajet(
        ville1 = trajet.ville2,
        ville2 = trajet.ville1,
        distance = trajet.distance
    ) 
    for trajet 
    in liste_trajets
]

liste_trajets.extend(liste_trajets_inverse)
liste_trajets
```

# Dictionnaire de distances

```python
dict_distances = {
    (obj_trajet.ville1, obj_trajet.ville2): obj_trajet.distance
    for obj_trajet
    in liste_trajets
}
dict_distances
```

# Répertorie toutes les villes

```python
set_villes = {
    ville
    for trajet
    in liste_trajets
    for ville in {trajet.ville1, trajet.ville2}
}
set_villes
```

```python
def func_get_distance_two_cities(
    in_villeA: str,
    in_villeB: str,
    in_liste_trajets: list[Trajet]
    ) -> int:
    """
    Calcule la distance entre deux villes 

    Args:
        in_villeA (str): ville de départ
        in_villeB (str): ville d'arrivée
        in_liste_trajets (list[Trajet]): liste des distances entre deux villes 

    Returns:
        int: distance entre les deux villes
    """
    
    for obj_trajet in in_liste_trajets:
        if (obj_trajet.ville1 == in_villeA) and (obj_trajet.ville2 == in_villeB):
            return obj_trajet.distance
```

```python
def func_get_distance_trajet(
    in_chemin: list[str],
    ):
    """Calcule la distance d'une suite de villes

    Args:
        in_chemin (list[str]): suite de villes 
    """
    
    distance_chemin = (
        sum(
            dict_distances[(villeA,villeB)]
            for villeA, villeB 
            in zip(in_chemin[:-1],in_chemin[1:])
        )
    )

    return distance_chemin
```

```python
# Pour chaque combo de villes possible
start = time.time()
shortest_distance = (
    min(
        func_get_distance_trajet(
            in_chemin = combinaison,
        )
        for combinaison 
        in permutations(set_villes)
    )
)

end = time.time()
print(end-start)

shortest_distance
```
