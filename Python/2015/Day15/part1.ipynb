# Import des librairies

```python
import re
from dataclasses import dataclass
```

# Definition classe

```python
@dataclass
class Ingredient:
    capacity: int
    durability: int
    flavor: int
    texture: int
    # calories: int
```

```python
@dataclass
class Receipt:
    capacity: int = 0
    durability: int = 0
    flavor: int = 0
    texture: int = 0
    # calories: int = 0
```

# Nettoyage des données

```python
def func_transform_text(in_row: str) -> Ingredient:

    # On spécifie le pattern a rechercher
    pattern = r"(\w+) (-?\d+)"

    # On cherche le pattern dans la ligne
    capacity, durability, flavor, texture, calories = re.findall(pattern,in_row)

    ingredient = Ingredient(
        capacity = int(capacity[1]),
        durability = int(durability[1]),
        flavor = int(flavor[1]),
        texture = int(texture[1]),
        # calories = int(calories[1])
    )
    
    return ingredient
```

# Lecture du fichier source

```python
with open('real.txt') as fichier:
    list_ingredient: list[Ingredient] = [func_transform_text(row) for row in fichier.read().split('\n')]

list_ingredient
```

# Traitement

```python
def func_score_receipt(
    in_teaspoons_ingredients: tuple[int,int]
) -> int:
    
    # Instancie une recette
    receipt = Receipt()

    # Pour chaque ingrédient, nombre de cuillère à café associé
    for ingredient, teaspoons_ingredient in zip(list_ingredient, in_teaspoons_ingredients):

        # On ajoute le score de l'ingredient aux features de la recette
        receipt.capacity += (ingredient.capacity * teaspoons_ingredient)
        receipt.durability += (ingredient.durability * teaspoons_ingredient)
        receipt.flavor += (ingredient.flavor * teaspoons_ingredient)
        receipt.texture += (ingredient.texture * teaspoons_ingredient)

    score_receipt = 1

    # On multiplie le score de chaque feature de la recette
    for feature_recette in receipt.__dict__.values():

        # Les features de la recette qui ont un score négatif sont ramenées à 0
        feature_recette = max(feature_recette,0)

        # Multiplie le score de chaque feature
        score_receipt *= feature_recette

    return score_receipt
```

# Génère toutes les combinaisons possibles

```python
list_combinaisons = []
for a in range(101):
    for b in range(101):
        for c in range(101):
            d = 100 - a - b - c
            if d < 0:
                break  
            list_combinaisons.append((a,b,c,d))


list_combinaisons
```

```python
max(
    func_score_receipt(
        in_teaspoons_ingredients = combinaisons
    )
    for combinaisons 
    in list_combinaisons
)
```
