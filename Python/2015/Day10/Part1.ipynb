# Version 1

```python
def func_convert(
    in_chain: str
):  

    last_digit = None
    streak = 1
    new_chain = []
    # print('chain : ',in_chain)

    # Pour chaque digit de la chaine
    for current_digit in in_chain:
        # print('current_digit : ',current_digit)
        # print('new_chain : ', new_chain)
        # print()

        # Si le digit actuel est le meme que le précédent
        if current_digit == last_digit:
            # print("meme digit que le precedent")
            
            # On incrémente le nombre de digits consécutifs
            streak += 1
        
        # Si le digit actuel est différent que le précédent
        else:
            
            # Le premier groupe sera ajouté lors de la prochaine itération pour ne pas que lat_digit=None
            # Pour les autres, on le fait maintenant
            if last_digit is not None:

                # On ajoute le combo précédent à la nouvelle chaine
                new_chain.extend([str(streak), last_digit])

            # On réinitilaise le nombre de digits consécutifs
            streak = 1
        
        last_digit = current_digit

    new_chain.extend([str(streak), last_digit])

            
    return ''.join(new_chain)
```

# Version 2

```python
def func_convert2(
    in_chain: list[int]
):  
    
    new_chain = []

    for digit, group in groupby(in_chain):
        new_chain.extend([len(list(group)),digit])
            
    return new_chain
```

```python
func_convert2([1,2,1,1])
```

```python
chain_digits = file2
for _ in range(40):
    chain_digits = func_convert2(chain_digits)
len(chain_digits)
```
