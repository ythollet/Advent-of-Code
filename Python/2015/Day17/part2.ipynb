# Nettoyage des données

```python
with open("real.txt") as file:
    list_containers = [int(e) for e in file.read().split("\n")]

liters = 150

list_containers
```

# Solution 2

```python
from itertools import combinations
from collections import defaultdict
```

```python
nb_possibilites = 0
dict_possibilités = defaultdict(int)

for r in range(len(list_containers)):
    combinaisons = combinations(list_containers, r)
    for combinaison in combinaisons:
        if sum(combinaison) == liters:
            # Recupère le nombre d'éléments dans le combo
            nb_elem_in_combinaison = len(combinaison)
            dict_possibilités[nb_elem_in_combinaison] += 1

# Récupère le nombre minimal d'élements par combo
min_elem_combo = min(dict_possibilités.keys())

dict_possibilités[min_elem_combo]
```
