# Nettoyage des données

```python
with open("real.txt") as file:
    list_containers = [int(e) for e in file.read().split("\n")]

liters = 150

list_containers
```

```python
# list_containers = [20, 15, 10, 5, 5]
# liters = 25
```

```python
def func_recursive(
    in_list_containers: list[int],
    current_index: int,
    current_sum: int,
    in_liters_target: int
):

    if current_sum == in_liters_target:
        return 1
    
    if current_sum > in_liters_target or current_index == len(in_list_containers):
        return 0
    
    current_container = in_list_containers[current_index]

    # Branche qui inclut le nombre actuel
    include = func_recursive(
        in_list_containers = in_list_containers,
        current_index = current_index + 1,
        current_sum = current_sum + current_container,
        in_liters_target = in_liters_target
    )

    exclude = func_recursive(
        in_list_containers = in_list_containers,
        current_index = current_index + 1,
        current_sum = current_sum,
        in_liters_target = in_liters_target
    )

    return include + exclude

func_recursive(
    in_list_containers = list_containers,
    in_liters_target = liters,
    current_index = 0,
    current_sum = 0,
)
```

```python
from itertools import combinations
```

```python
nb_possibilites = 0

for r in range(len(list_containers)):
    combinaisons = combinations(list_containers, r)
    for combinaison in combinaisons:
        if sum(combinaison) == liters:
            nb_possibilites += 1

nb_possibilites
```
