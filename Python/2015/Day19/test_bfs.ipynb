# Solution finale: BFS

```python
from collections import deque

# Graphe représentant l'école
graph = {
    "Classe": ["Couloir", "Cour"],
    "Couloir": ["Classe", "Bibliothèque"],
    "Cour": ["Classe", "Cantine"],
    "Bibliothèque": ["Couloir", "Salle de sport"],
    "Cantine": ["Cour"],
    "Salle de sport": ["Bibliothèque"]
}
```

```python
def bfs(in_graph, start, steps):

    visited = set()
    pile = deque([(start,0)])

    while pile:

        steps += 1
        piece, steps = pile.popleft()

        print(piece)

        if piece == 'Salle de sport':
            return steps
        
        possibilities = in_graph[piece]

        for possibility in possibilities:
            if possibility not in visited:
                pile.append((possibility,steps+1))
                visited.add(possibility)
```

```python
bfs(graph, 'Classe', 0)
```
