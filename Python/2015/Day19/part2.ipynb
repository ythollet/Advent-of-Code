```python
with open("real.txt") as file:
    data = [row for row in file.read().splitlines()]

transformations = data[:-2]
target_sentence = data[-1]

list_sentence = list(target_sentence)

transformations, target_sentence
```

```python
dict_transform = {}

# Pour chacune des lignes de transformations
for row in transformations:

    # On récupère la valeur originale et sa conversion
    original, new = row.split(' => ')

    # On initialise le Dictionnaire de transformations
    dict_transform.setdefault(original,[]).append(new)

dict_transform
```

```python
dict_letters_sentence = defaultdict(int)
# Compter le nombre de fois que chaque lettre apparait dans la phrase et stocker le res dans le dict

for letter in target_sentence:
    dict_letters_sentence[letter] += 1

dict_letters_sentence
```

```python
def func_get_all_possibilities_for_sentence(
    in_dict_transform: dict,
    in_sentence: str    
) -> list[str]:
    
    list_sentence = list(in_sentence)
    set_possibilities = set()

    # Pour chaque mollécule transfromable
    for key_transform in in_dict_transform:

        # print(key_transform)

        # Parcours la phrase a la recherche d'une mollécule
        for index_potential_mollecule in range(0, len(in_sentence)-len(key_transform)+1):

            # Applique la zone de recherche actuelle sur la phrase
            potential_mollecule = in_sentence[index_potential_mollecule:index_potential_mollecule + len(key_transform)]

            # A chaque fois qu'on trouve une mollécule, on la transforme
            if potential_mollecule == key_transform:
                
                # La mollécule potentielle devient une mollécule
                mollecule = potential_mollecule

                # Conversion de la mollecule en une nouvelle mollécule
                list_possibilities_transformations_mollecule = in_dict_transform[mollecule]

                # Pour chaque possibilité de transformation de la mollécule
                for possibility_transformation_mollecule in list_possibilities_transformations_mollecule:

                    # Nouvelle phrase: partie avant la mollecule + nouvelle mollécule + partie après la mollécule
                    new_list_sentence = list_sentence[:index_potential_mollecule] + [possibility_transformation_mollecule] + list_sentence[index_potential_mollecule + len(key_transform):]

                    # Conversion de la nouvelle phrase (list[str] -> str)
                    new_sentence = ''.join(new_list_sentence)

                    # Ajouter la nouvelle phrase au set
                    set_possibilities.add(new_sentence)
 
    return list(set_possibilities)
```

On part du target pour arriver à "e"

```python
def func_get_all_reversed_possibilities_for_sentence(
    in_reversed_dict_transform: dict,
    in_sentence: str    
) -> list[str]:
    
    set_possibilities = set()
    list_sentence = list(in_sentence)


    # Pour chaque mollécule transfromable
    for key_transform in in_reversed_dict_transform:

        # print(key_transform)

        # Parcours la phrase a la recherche d'une mollécule
        for index_potential_mollecule in range(0, len(in_sentence)-len(key_transform)+1):

            # Applique la zone de recherche actuelle sur la phrase
            potential_mollecule = in_sentence[index_potential_mollecule:index_potential_mollecule + len(key_transform)]

            # A chaque fois qu'on trouve une mollécule, on la transforme
            if potential_mollecule == key_transform:
                
                # La mollécule potentielle devient une mollécule
                mollecule = potential_mollecule

                # Conversion de la mollecule en une nouvelle mollécule
                list_possibilities_transformations_mollecule = in_reversed_dict_transform[mollecule]

                # Pour chaque possibilité de transformation de la mollécule
                for possibility_transformation_mollecule in list_possibilities_transformations_mollecule:

                    # Nouvelle phrase: partie avant la mollecule + nouvelle mollécule + partie après la mollécule
                    new_list_sentence = list_sentence[:index_potential_mollecule] + [possibility_transformation_mollecule] + list_sentence[index_potential_mollecule + len(key_transform):]

                    # Conversion de la nouvelle phrase (list[str] -> str)
                    new_sentence = ''.join(new_list_sentence)

                    # Ajouter la nouvelle phrase au set
                    set_possibilities.add(new_sentence)
 
    return list(set_possibilities)
```

```python
# On commence par inverser le dictionnaire de trasformation
reversed_dict_transform = {}

for key_dict_transform, v_dict_transform in dict_transform.items():
    for value in v_dict_transform:
        reversed_dict_transform.setdefault(value,[]).append(key_dict_transform)


reversed_dict_transform
```

```python
def func_recursive_inverse(
    in_current_sentence: str,
    in_nb_steps: int,
    in_reversed_dict_transform: dict,
):
    # print("STEP ",in_nb_steps)
    # print("sentence", in_current_sentence)

    global lower_steps

    new_sentences = func_get_all_reversed_possibilities_for_sentence(
        in_reversed_dict_transform= in_reversed_dict_transform,
        in_sentence = in_current_sentence
    )
    
    for new_sentence in new_sentences:
        # print(new_sentence)
        # print()
        print(new_sentence)

        if new_sentence == "e":
            if in_nb_steps <= lower_steps:
                lower_steps = in_nb_steps 
                return lower_steps
        
        else:
            func_recursive_inverse(
                in_current_sentence = new_sentence,
                in_nb_steps = in_nb_steps+1,
                in_reversed_dict_transform = in_reversed_dict_transform,
            )
    return lower_steps
```

# Solution finale: BFS

```python
start = target_sentence # On part du resultat
target = "e"
nb_steps = 0
lower_steps=float('+inf')

visited = set([start])

pile = deque([(start, 0)])

while pile:

    # Si la nouvelle phrase qu'on regarde a nécessité trop d'étapes
    if nb_steps >= lower_steps:
        continue
    print(pile)
    
    current_sentence, nb_steps = pile.popleft()

    if current_sentence == target:
        print(nb_steps)
        lower_steps = nb_steps
        

    new_sentences = func_get_all_reversed_possibilities_for_sentence(
        in_reversed_dict_transform = reversed_dict_transform,
        in_sentence = current_sentence
    )
    
    for new_sentence in new_sentences:

        if new_sentence not in visited:
            pile.append((new_sentence,nb_steps+1))
            visited.add(new_sentence)
lower_steps
```
