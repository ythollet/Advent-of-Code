# Lecture du fichier source

```python
with open('real.txt') as fichier:
    file = [
        tuple(map(int,numbers.split('x'))) # largeur, longueur, hauteur
        for numbers 
        in fichier.read().split('\n')
    ]
file
```

# Traitement

```python
total_wrapping_paper = 0

# Pour chaque dimensions de cadeau
for l,L,h in file:

    permieter_face1 = 2*(l+L)
    permieter_face2 = 2*(L+h)
    permieter_face3 = 2*(h+l)

    smallest_perimeter = min(permieter_face1, permieter_face2, permieter_face3)
    # print('smallest_perimeter',smallest_perimeter)
    volume = l*L*h
    # print('noeud',volume)

    current_present_wrapping_paper = volume + min((permieter_face1, permieter_face2, permieter_face3))
    total_wrapping_paper += current_present_wrapping_paper
    
print(total_wrapping_paper)
```
