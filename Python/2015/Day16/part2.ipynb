# Nettoyage des données

```python
def func_clean_row(in_row: str):

    pattern = r"(\w+)"
    res = re.findall(pattern, in_row)[2:]

    dict_sue = {}

    for cle, valeur in zip(res[::2], res[1::2]): 
        dict_sue[cle] = int(valeur)

    return dict_sue
    
```

```python
with open("test.txt") as file:
    data = [func_clean_row(row) for row in file.read().split("\n")]
data
```

```python
dict_real_sue = {
    'children': 3,
    'cats': 7,
    'samoyeds': 2,
    'pomeranians': 3,
    'akitas': 0,
    'vizslas': 0,
    'goldfish': 5,
    'trees': 3,
    'cars': 2,
    'perfumes': 1,
}
```

```python
# Pour chacune des 500 tantes
for num_sue, dict_sue in enumerate(data):
    is_real_sue = True

    # On regarde chacune de ses caractéristiques
    for key_current_sue, value_current_sue in dict_sue.items():

        # Si la caractéritique est la même que celle de la vraie tante
        value_real_sue = dict_real_sue[key_current_sue]

        # Vérif du nombre min de "cats" et de "trees"
        if key_current_sue in ['cats', 'trees']:
            if value_current_sue <= value_real_sue:
                is_real_sue = False
                break
        
        # Vérif du nombre maximal de "pomeranians" et de "goldfish"
        if key_current_sue in ['pomeranians', 'goldfish']:
            if value_current_sue >= value_real_sue:
                is_real_sue = False
                break

        if key_current_sue not in ['cats', 'trees', 'pomeranians', 'goldfish']:
            if value_real_sue != value_current_sue:
                is_real_sue = False
                break

    if is_real_sue is True:
        print(num_sue+1)
        print(dict_sue)
        print()

            
```
