# Import des librairies

```python
from dataclasses import dataclass
```

# Classe

```python
@dataclass
class Instruction:
    direction: str
    unities: int
```

```python
@dataclass
class Position:
    x: int
    y: int
    direction: str
```

# Lecture du fichier source

```python
with open("real.txt") as fichier:
    list_instructions = [
        Instruction(
            direction = instruction[0],
            unities = int(instruction[1:])
        ) 
        for instruction 
        in fichier.read().split(', ')
    ]

list_instructions
```

```python
# dict_directions[direction_avant_instruction][direction_instruction][axis]
dict_directions = {
    'N': {
        'R':{
            'x':1,
            'y':0,
            'dir':'E'
        },
        'L':{
            'x':-1,
            'y':0,
            'dir':'O'
        }
    },
    'S':{
        'R':{
            'x':-1,
            'y':0,
            'dir':'O'
        },
        'L':{
            'x':1,
            'y':0,
            'dir':'E'
        }
    },
    'E':{
        'R':{
            'x':0,
            'y':-1,
            'dir':'S'
        },
        'L':{
            'x':0,
            'y':1,
            'dir':'N'
        }
    },
    'O':{
        'R':{
            'x':0,
            'y':1,
            'dir':'N'
        },
        'L':{
            'x':0,
            'y':-1,
            'dir':'S'
        }
    }
}
```

```python
def func_avancer(
    in_instruction: Instruction,
    in_current_pos: Position,
    in_dict_directions: dict,
    in_visited: set
) -> Position:
    
    delta_x, delta_y, new_dir = in_dict_directions[in_current_pos.direction][in_instruction.direction].values()
    
    for unity in range(1,in_instruction.unities+1):

        new_pos = Position(
            direction = new_dir,
            x = in_current_pos.x + delta_x * unity,
            y = in_current_pos.y + delta_y * unity
        )

        coord = (new_pos.x, new_pos.y)

        if coord in in_visited:
            print("deja visité")
            print(new_pos)
            print(abs(new_pos.x) + abs(new_pos.y))
            return
        else:
            in_visited.add(coord)
    return new_pos
```

```python
pos = Position(
    x=0,
    y=0,
    direction='N'
)

visited = {(pos.x, pos.y)}
```

```python
for instruction in list_instructions:
    try:
        pos = func_avancer(
            in_current_pos = pos,
            in_dict_directions = dict_directions,
            in_instruction = instruction,
            in_visited = visited
        )
    except AttributeError as e:
        break
```
