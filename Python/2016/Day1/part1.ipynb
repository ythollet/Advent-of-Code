# Import des librairies

```python
from dataclasses import dataclass
```

# Classe

```python
@dataclass
class Instruction:
    direction: str
    unities: int
```

```python
@dataclass
class Position:
    x: int
    y: int
    direction: str
```

# Lecture du fichier source

```python
with open("real.txt") as fichier:
    list_instructions = [
        Instruction(
            direction = instruction[0],
            unities = int(instruction[1:])
        ) 
        for instruction 
        in fichier.read().split(', ')
    ]

list_instructions
```

```python
# dict_directions[direction_avant_instruction][direction_instruction][axis]
dict_directions = {
    'N': {
        'R':{
            'x':1,
            'y':0,
            'dir':'E'
        },
        'L':{
            'x':-1,
            'y':0,
            'dir':'O'
        }
    },
    'S':{
        'R':{
            'x':-1,
            'y':0,
            'dir':'O'
        },
        'L':{
            'x':1,
            'y':0,
            'dir':'E'
        }
    },
    'E':{
        'R':{
            'x':0,
            'y':-1,
            'dir':'S'
        },
        'L':{
            'x':0,
            'y':1,
            'dir':'N'
        }
    },
    'O':{
        'R':{
            'x':0,
            'y':1,
            'dir':'N'
        },
        'L':{
            'x':0,
            'y':-1,
            'dir':'S'
        }
    }
}
```

```python
def func_avancer(
    in_instruction: Instruction,
    in_current_pos: Position,
    in_dict_directions: dict
) -> Position:
    
    delta_x, delta_y, new_dir = in_dict_directions[in_current_pos.direction][in_instruction.direction].values()

    new_pos = Position(
        direction = new_dir,
        x = in_current_pos.x + delta_x * in_instruction.unities,
        y = in_current_pos.y + delta_y * in_instruction.unities
    )
    return new_pos
```

```python
pos = Position(
    x=0,
    y=0,
    direction='N'
)
```

```python
for instruction in list_instructions:
    pos = func_avancer(
        in_current_pos = pos,
        in_dict_directions = dict_directions,
        in_instruction = instruction
    )


pos
```
