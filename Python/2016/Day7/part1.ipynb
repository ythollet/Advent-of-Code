# Import des librairies

```python
from dataclasses import dataclass
import re 
```

# Classe

```python
@dataclass
class Ip:
    out_brackets: list[str]
    in_brackets: list[str]
```

```python
def func_decode_row(
    in_row: str
) -> Ip:
    
    pattern = r"[\[\]]"
    parts = re.split(pattern, in_row)

    out_brackets = [parts[num_part] for num_part in range(0,len(parts),2)]
    in_brackets = [parts[num_part] for num_part in range(1,len(parts),2)]

    ip = Ip(
        out_brackets = out_brackets,
        in_brackets = in_brackets
    )

    return ip
```

# Lecture du fichier source

```python
with open("real.txt") as fichier:
    list_ip = [
        func_decode_row(row)
        for row
        in fichier.read().splitlines()
    ]
    
list_ip
```

```python
def func_verif_reversed_pair(
    in_chain: str
) -> bool:
    
    length_chain = len(in_chain)
    range_start = range(0, length_chain-3)
    range_end = range(4, length_chain+1)

    for start_range_search, end_range_search in zip(range_start, range_end):
        scope = in_chain[start_range_search:end_range_search]

        if (scope[0] == scope[3]) and (scope[1] == scope[2]) and (scope[2]!=scope[3]):
            return True
        
    return False
```

```python
def func_verif_supports_tls(
    in_ip: Ip
):
    
    for in_brackets in in_ip.in_brackets:
        if func_verif_reversed_pair(in_brackets):
            return False
        
    for out_bracket in in_ip.out_brackets:
        if func_verif_reversed_pair(out_bracket):
            print(in_ip, " is OK")
            return True
    
    return False
```

```python
sum(func_verif_supports_tls(ip) for ip in list_ip)
```
