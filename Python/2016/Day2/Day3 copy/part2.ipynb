# Lecture du fichier source

```python
data = []

with open("real.txt") as fichier:

    data = np.array([
        [int(side) for side in triangle.split()]
        for triangle 
        in fichier.read().splitlines()
    ])

data
```

```python
list_triangle_ordered = []

for num_col in range(3):
    for start_row, end_row in zip(range(0, len(data)+1, 3), range(3, len(data)+1, 3)):
        triangle = (data[start_row:end_row,num_col].tolist())
        list_triangle_ordered.append(triangle)
list_triangle_ordered
```

```python
def func_verif_triangle(
    in_triangle: list[int]
) -> bool:

    for num_side in range(3):
        side1 = in_triangle[num_side]
        side2 = in_triangle[(num_side+1)%3]
        side3 = in_triangle[(num_side+2)%3]

        if side1 + side2 <= side3:
            return False
        
    return True
```

```python
nb_triangles = sum(
    func_verif_triangle(triangle)
    for triangle 
    in list_triangle_ordered
    )

nb_triangles
```
