# Import des librairies

```python
from dataclasses import dataclass
import numpy as np
```

# Classe

```python
@dataclass
class Position:
    row: int
    col: int
```

# Lecture du fichier source

```python
with open("real.txt") as fichier:
    list_instructions = [
        [i for i in instruction]
        for instruction 
        in fichier.read().splitlines()
    ]

list_instructions
```

# Dictionnaire de déplacements

```python
dict_moove = {
    'U': {
        'row': -1,
        'col': 0
        },
    'D': {
        'row': 1,
        'col': 0
        },
    'R': {
        'row': 0,
        'col': 1
        },
    'L': {
        'row': 0,
        'col': -1
        },   
}
```

# Matrice pavé numérique

```python
matrix = np.array([[3*num_row + row for row in range(1,4)] for num_row in range(3)])
matrix
```

# Traitement

```python
current_pos = Position(
    row = 1,
    col = 1
)
```

```python
def func_get_new_pos(
    in_letter_instruction: str,
    in_current_pos: Position
):
    
    # Initialise la nouvelle position
    new_pos = Position(
        row = in_current_pos.row + dict_moove[in_letter_instruction]['row'],
        col = in_current_pos.col + dict_moove[in_letter_instruction]['col']
    )

    # Si la position reste dans les limites de la matrice
    if (0 <= new_pos.row <= 2) and (0 <= new_pos.col <= 2):

        # On retourne la nouvelle position
        return new_pos
    
    # Si la position dépasse les limites de la matrice
    else:

        # On ne bouge pas --> on retourne l'ancienne position
        return in_current_pos
```

```python
def func_get_code_for_row(
    in_row_instruction: list[str],
    in_current_pos: Position
) -> Position:
    
    current_pos = in_current_pos

    for letter_instruction in in_row_instruction:

        print('current_pos : ',in_current_pos)
        new_pos = func_get_new_pos(
            in_letter_instruction = letter_instruction,
            in_current_pos = current_pos
        )

        current_pos = new_pos
    return new_pos
```

```python
code = []

# Pour chaque ligne d'instruction
for row_instruction in list_instructions:

    # On détermine la nouvelle position
    new_pos = func_get_code_for_row(
        in_current_pos = current_pos,
        in_row_instruction = row_instruction
    )

    # On détermine la valeur de cette position dans la matrice du pavé numérique
    value_keypad = matrix[new_pos.row,new_pos.col]

    # On stocke la valeur du pavé numérique correspondante
    code.append(value_keypad)

    # On met à jour la position actuelle
    current_pos = new_pos

''.join([str(c) for c in code])
```
