```python
def func_str_to_hexadecimal(
    in_chain: str
) -> str:

    # Conversion de la chaine en binaire
    binary_door_id = in_chain.encode()

    # Créeation d'un objet MD5
    md5_obj = md5(binary_door_id)

    # Converion de l'objet md5 en chaine Hexadécimale
    hexadecimal_door_id = md5_obj.hexdigest()

    # Renvoie la chaine hexadécimale
    return hexadecimal_door_id
```

```python
def func_get_next_char_password(
    in_door_id: str,
    in_i: str
) -> int:

    # Tant que les 5 premiers caractèrers de la chaine ne sont pas '00000'  
    while True:

        # Ajoute le chiffre à la fin de la chaine 
        chain = in_door_id + str(in_i)

        # Convertit en hexadécimal
        chain_hexa = func_str_to_hexadecimal(chain)
        
        # Si les les 5 premiers caractèrers de la chaine sont '00000'  
        if chain_hexa[:5] == '00000':

            # On retourne le nombre d'itérations
            return (in_i, chain_hexa[5])
        
        # Sinon on incrémente i
        in_i += 1
```

```python
res_i = -1
password = []
for nb_char_password in range(8):
    res_i, res_char_password = func_get_next_char_password(
        in_door_id = door_id,
        in_i = res_i+1
    )

    password.append(res_char_password)

password
```
