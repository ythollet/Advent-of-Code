# Import des librairies

```python
from dataclasses import dataclass
import re
import numpy as np 
```

# Classe

```python
@dataclass
class Rect:
    cols: int
    rows: int
```

```python
@dataclass
class Rotate:
    axis: str
    index_axis: int
    decallage: int
    _end: int = None

    @property
    def end(self):
        if self._end is None:
            self._end = self.index_axis + self.decallage
        return self._end
```

```python
def func_decode_row(
    in_row: str
) -> Rect | Rotate:
    
    if in_row.startswith('rect'):
        pattern = r"rect (\d+)x(\d+)"
        cols, rows = re.match(pattern, in_row).groups()
        
        rect = Rect(
            cols = int(cols),
            rows = int(rows)
        )
        
        return rect
    
    else:
        pattern = r"rotate (\w+) \w+=(\d+) by (\d+)"
        axis, start, decallage = re.match(pattern, in_row).groups()

        rotate = Rotate(
            axis = axis,
            index_axis = int(start),
            decallage = int(decallage)
        )
        return rotate
```

# Lecture du fichier source

```python
with open("real.txt") as fichier:
    list_instructions = [
        func_decode_row(row)
        for row
        in fichier.read().splitlines()
    ]
    
list_instructions
```

```python
wide_matrix = 50
tall_matrix = 6

matrix = np.array([
    ['.']*wide_matrix
    for row in range(tall_matrix)
])
matrix
```

```python
# Pour chaque instruction
for instruction in list_instructions:
    
    # Si c'est une creation de Rectangle
    if isinstance(instruction, Rect):

        rect = instruction
        
        matrix[0:rect.rows, 0:rect.cols] = "#"


    # Si c'est une rotation
    else:
        rotate = instruction

        # Si le décallage se fait sur une ligne
        if rotate.axis == 'row':
            
            # Endroit où l'on va split la ligne en 2
            index_coupure = wide_matrix-rotate.decallage

            # On récupère les n derniers élements de la lignes (n correspond à la taille du décallage)
            right_part = matrix[rotate.index_axis, index_coupure:]

            # On récupère l'autre partie
            left_part = matrix[rotate.index_axis, :index_coupure]

            # On crée la nouvelle ligne
            new_row = np.concatenate([right_part, left_part])

            # On insère la nouvelle ligne dans la matrice
            matrix[rotate.index_axis, :] = new_row

        # Si le décallage se fait sur une colonne
        else:

            # Endroit où l'on va split la colonne en 2
            index_coupure = tall_matrix-rotate.decallage

            # On récupère les n derniers élements de la colonne (n correspond à la taille du décallage)
            up_part = matrix[:index_coupure, rotate.index_axis]

            # On récupère l'autre partie
            down_part = matrix[index_coupure:, rotate.index_axis]

            # On crée la nouvelle colonne
            new_col = np.concatenate([down_part, up_part], axis=0)

            # On insère la nouvelle colonne dans la matrice
            matrix[:, rotate.index_axis] = new_col

```
