# Import des librairies

```python
from dataclasses import dataclass
import re
import string
```

# Classe

```python
@dataclass
class Room:
    name: str
    id: int
    checksum: str
```

# Initialise une Room

```python
def func_init_room(
    in_row: str
) -> Room:
    
    pattern = r"([a-z-]+)-(\d+)\[([a-z]+)\]"

    name, id, checksum = re.search(pattern, in_row).groups()

    room = Room(
        name = name.replace('-',''),
        id = int(id),
        checksum = checksum 
    )

    return room
```

# Lecture du fichier source

```python
with open("real.txt") as fichier:
    list_rooms = [
    func_init_room(in_row=row)
    for row 
    in fichier.read().splitlines()
]
list_rooms
```

```python
def func_is_real_room(
    in_room: Room
) -> bool:
    
    dict_count_letters = {}

    # Compte les lettres contenues dans le name de la room
    for letter in in_room.name:
        dict_count_letters.setdefault(letter, 0)
        dict_count_letters[letter] += 1
    
    # Trie les lettres par ordre décroissant de fréquence d'apparition
    list_tuple_sorted = sorted(
        dict_count_letters.items(),
        key = lambda item: (-item[1], item[0]),
    )

    # On garde que les 5 lettres les plus fréquentes
    list_tuple_sorted = list_tuple_sorted[:5]

    # Transforme la liste de tuple en liste de str
    list_letters_sorted = [
        letter 
        for letter, freq
        in list_tuple_sorted
    ]

    chain_letters_sorted = ''.join(list_letters_sorted)

    if chain_letters_sorted == in_room.checksum:
        return True
    
    else:
        return False
```

```python
sum_id = sum(
    room.id
    for room in list_rooms
    if func_is_real_room(in_room = room)
)
sum_id
```
