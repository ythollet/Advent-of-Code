# Lecture du fichier source

```python
with open("real.txt") as fichier:
    matrix = np.array([
        [cell for cell in row]
        for row
        in fichier.read().splitlines()
    ])
    


n_cols_matrix = len(matrix[0,:])

matrix
```

```python
def func_get_most_common_letter(
    in_num_col: int
) -> str:

    letters_column = matrix[:,in_num_col].tolist()

    compteur = Counter(letters_column)
    
    return min(compteur, key=compteur.get)
```

```python
''.join([func_get_most_common_letter(in_num_col=num_col) for num_col in range(n_cols_matrix)])
```
