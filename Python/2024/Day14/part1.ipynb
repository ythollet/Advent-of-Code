# Import des librairies

```python
from collections import defaultdict
```

# Lecture du fichier

```python
with open("real.txt") as fichier:
    file = fichier.read().strip().split("\n")
file
```

# Praparation des données

```python
list_robots = []

# Pour chaque robot
for robot in file:

    # On split ses caractéristiques
    details_robot = robot.split(' ')

    # On récupère les valeurs de p et v (sans p= ou v=)
    p = tuple(map(int,details_robot[0][2:].split(',')))
    v = tuple(map(int,details_robot[1][2:].split(',')))

    # On met les caractéristiques du robot dans un dict
    dict_robot = {
        'p' : p,
        'v' : v
    }

    # On met le dict dans une liste commune
    list_robots.append(dict_robot)
list_robots

x_max = 101
y_max = 103

mid_x = (x_max -1) // 2
mid_y = (y_max -1) // 2

mid_x
mid_y
```

# Traitement

```python
dict_final_pos = defaultdict(int)
top_left = top_right = down_left = down_right = 0
for dict_robot in list_robots:
    x_start, y_start = dict_robot['p']
    x_moove, y_moove = dict_robot['v']
    final_pos = ((x_start + 100 * x_moove) % x_max, (y_start + 100 * y_moove) % y_max)

    # Si la nouvelle pos est à gauche
    if final_pos[0] < mid_x:

        # Si la nouvelle pos est en haut à gauche
        if final_pos[1] < mid_y:
            top_left += 1

        # Si la nouvelle pos est en bas à gauche
        elif final_pos[1] > mid_y:
            down_left += 1
    
    # Si la nouvelle pos est à droite
    elif final_pos[0] > mid_x:

        # Si la nouvelle pos est en haut à droite
        if final_pos[1] < mid_y:
            top_right += 1

        # Si la nouvelle pos est en bas à droite
        elif final_pos[1] > mid_y:
            down_right += 1


    dict_final_pos[final_pos] +=1
dict_final_pos

score = (top_left * top_right * down_left *down_right)
score
```
