# Import des librairies

```python
import numpy as np
```

# Lecture du fichier source

```python
with open ("real.txt") as fichier:
    grille = fichier.readlines()
grille
```

# Nettoyage du fichier

```python
# On récupère la position (horizontale et verticale) et la direction de l'agent au début
pos_v_agent, pos_h_agent = next((num_line,line.index("^")) for num_line,line in enumerate(grille) if "^" in line)  # next permet de s'atrreter quand on trouve "^"
# On historise les positions de l'agent
historique_positions=[(pos_v_agent, pos_h_agent)]
direction_agent = "haut" # direction du regard de l'agent

# Maitenant qu'on a identifié la position de l'agent on supprime le "^"
grille = [line.replace("^",".") for line in grille]

[line.replace("\n","") for line in grille]
```

# Traitement

```python
# fonction qui détermine si l'agent peut avancer, si oui, on avance, sinon on tourne et on avance
def func_avancer(var_pos_v_agent:int,
                 var_pos_h_agent:int,
                 var_direction_agent:str,
                 var_grille:list[str]
                )->tuple[bool, int, int, str] | None:
    
    # On définit un dictionnaire qui associe chaque direction à un déplacement dans la grille
    dico_moove={"haut":   (-1, 0),
                "droite": (0, 1),
                "bas":    (1, 0),
                "gauche": (0, -1)
                }
    
    list_directions = ["haut","droite","bas","gauche"]

    limit_v_grille = len(var_grille)-1
    limit_h_grille = len(var_grille[0])-1
    
    # On définit la nouvelle position si on continue dans la même direction (new pos = ancinne position + déplacement)
    new_pos_v_agent = var_pos_v_agent + dico_moove[var_direction_agent][0]  # Déplacement vertical
    new_pos_h_agent = var_pos_h_agent + dico_moove[var_direction_agent][1]  # Déplacement horizontal

    # Si on dépasse de la grille on s'arrete
    if not (0<=new_pos_v_agent<=limit_v_grille and 0<=new_pos_h_agent<=limit_h_grille):
        return
    # Si l'agent ne plus avancer dans la même direction
    if var_grille[new_pos_v_agent][new_pos_h_agent] == "#":
        continuer = False
        # On recupere l'indice de la nouvelle direction dans la liste ["haut","droite","bas","gauche"]
        index_new_direction_agent = (list_directions.index(var_direction_agent)+1)%4
        
        # On définit la nouvelle direction
        new_direction_agent = list_directions[index_new_direction_agent]

        # On choisit un autre déplacement 
        new_pos_v_agent = var_pos_v_agent + dico_moove[new_direction_agent][0]  # Déplacement vertical
        new_pos_h_agent = var_pos_h_agent + dico_moove[new_direction_agent][1]  # Déplacement horizontal

    
        

    # Si l'agent peut encore avancer dans la même direction
    else:
        continuer = True
        # On définit la nouvelle direction (qui reste la même dans ce cas)
        new_direction_agent = var_direction_agent
    print(f"Ancienne direction : {var_direction_agent}")
    print(f"Continuer dans la même direction: {continuer}")
    print(f"Nouvelle direction : {new_direction_agent}")
        
    print("\n")
        
    # On relance la fonction    
    return continuer, new_pos_v_agent, new_pos_h_agent, new_direction_agent      
    




# On compte le nombre de déplacement de l'agent
res_func = ""
while res_func is not None:
    print(f"nouvelle position : {pos_v_agent,pos_h_agent}")
    res_func = func_avancer(pos_v_agent, pos_h_agent, direction_agent, grille)
    if res_func is not None:
        continuer, pos_v_agent, pos_h_agent, direction_agent = res_func
        pos_agent=(pos_v_agent,pos_h_agent)
        if pos_agent not in historique_positions:
            historique_positions.append(pos_agent)

historique_positions
        
        
```
