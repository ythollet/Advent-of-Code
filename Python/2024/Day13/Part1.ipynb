# Import des librairies

```python
import re
from sympy import symbols, Eq, solve, diophantine
```

# Lecture du fichier

```python
with open("test.txt") as fichier:
    # On sépares les machines
    file = fichier.read().strip().replace(' ','').split("\n\n")

file
```

# Préparation des données

```python
list_machines = []

# Pour chaque machine
for machine in file:
    machine_features = {}
    # Pour les détails de la machine
    for features in machine.split("\n"):
        # On split chaque détail
        list_features = re.split("[:|, ]",features)
        # On crée un dictionnaire pour chaque détail
        dict_feature = {list_features[0] : (int(list_features[1][2:]), int(list_features[2][2:]))}
        # On fusionne les détails dans un dict commun
        machine_features |= dict_feature
    list_machines.append(machine_features)


list_machines
```

```python
def func_solve_dict(in_dict_machine : dict[str : (int,int)]):
    """ Prends en parametre les infos d'une machine et retourne le nombre de token pour la machine  """
    print(in_dict_machine)
    x, y = symbols('x y', integer=True)
    eq1 = Eq(
            (in_dict_machine['ButtonA'][0] * x) + 
            (in_dict_machine['ButtonB'][0] * y) ,
            in_dict_machine['Prize'][0] 
        )
    eq2 = Eq(
            (in_dict_machine['ButtonA'][1] * x) + 
            (in_dict_machine['ButtonB'][1] * y) ,
            in_dict_machine['Prize'][1] 
        )
    
    list_res = solve((eq1, eq2), (x, y), dict=True)
    if list_res == []:
        return 0
    print(list_res)

    return 3 * list_res[0][x] + list_res[0][y]
    
```

```python
# On calucle la somme du nombre min de token pour toutes les machines
sum(func_solve_dict(in_dict_machine = dict_machine) for dict_machine in list_machines)
```
