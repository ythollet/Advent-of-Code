```python
_,program_str =  file[1][0].split(' ')
program = [int(number) for number in program_str.split(',')]

registers_str = [register.replace('Register ','') for register in file[0]]
registers = {register[0] : int(register[2:]) for register in registers_str}
```

# Fonctions

```python
def func_init_operand(in_lit_operand):
    
    # Combo operand peut etre soit la valeur littérale soit la valeur d'un registre
    if in_lit_operand == 4:
        combo_operand = registers['A']
    elif in_lit_operand == 5:
        combo_operand = registers['B']
    elif in_lit_operand == 6:
        combo_operand = registers['C']
    else:
        combo_operand = in_lit_operand

    return combo_operand
```

```python
def func_opcode_0(in_registers, in_combo_operand):
    A_value = in_registers['A']
    in_registers['A'] = A_value //(2**in_combo_operand)
    return in_registers
```

```python
def func_opcode_1(in_registers, in_lit_operand):
    B_value = in_registers['B']

    res = B_value ^ in_lit_operand
    in_registers['B'] = res
    return in_registers
```

```python
def func_opcode_2(in_registers, in_combo_operand):

    in_registers['B'] = in_combo_operand % 8
    return in_registers
```

```python
def func_opcode_3(in_registers, in_lit_operand, in_index_pointer):
    # Si le registre A vaut 0, on ne fais rien
    if in_registers['A'] == 0:
        in_index_pointer += 2
        return in_index_pointer
    
    # Sinon on change le pointeur
    new_index_pointer = in_lit_operand


    return new_index_pointer
```

```python
def func_opcode_4(in_registers, in_lit_operand):
    B_value = in_registers['B']
    C_value = in_registers['C']

    res = B_value ^ C_value
    in_registers['B'] = res
    return in_registers
```

```python
def func_opcode_5(in_registers, in_combo_operand):

    return (in_combo_operand % 8)
```

```python
def func_opcode_6(in_registers, in_combo_operand):
    A_value = in_registers['A']
    in_registers['B'] = A_value //(2**in_combo_operand)
    return in_registers
```

```python
def func_opcode_7(in_registers, in_combo_operand):
    A_value = in_registers['A']
    in_registers['C'] = A_value //(2**in_combo_operand)
    return in_registers
```

```python
# Initialisation
index_pointer = 0
list_out = []

while index_pointer <= (len(program)-1):
    # On récupère l'opcode et l'operand
    opcode = program[index_pointer]
    lit_operand = program[index_pointer+1]

    # On initialise le combo operand
    combo_operand = func_init_operand(in_lit_operand = lit_operand)

    if opcode == 0:
        registers = func_opcode_0(in_registers = registers,
                                  in_combo_operand = combo_operand)
        
    elif opcode == 1:
        registers = func_opcode_1(in_registers = registers,
                                  in_lit_operand = lit_operand)
        
    elif opcode == 2:
        registers = func_opcode_2(in_registers = registers,
                                  in_combo_operand = combo_operand)
        
    elif opcode == 3:
        index_pointer = func_opcode_3(in_registers = registers,
                                      in_lit_operand = lit_operand,
                                      in_index_pointer = index_pointer)
    elif opcode == 4:
        registers = func_opcode_4(in_registers = registers,
                                  in_lit_operand = lit_operand)
        
    elif opcode == 5:
        res_opcode_5 = func_opcode_5(in_registers = registers,
                      in_combo_operand = combo_operand)
        list_out.append(res_opcode_5)
        
    elif opcode == 6:
        registers = func_opcode_6(in_registers = registers,
                                  in_combo_operand = combo_operand)
        
    elif opcode == 7:
        registers = func_opcode_7(in_registers = registers,
                                  in_combo_operand = combo_operand)        
    
    if opcode != 3:
        index_pointer += 2
list_out_join = ','.join(list_out)
list_out_join
```
