# Fonctions

```python
def func_init_operand(in_lit_operand, in_A, in_B, in_C):
    
    # Combo operand peut etre soit la valeur littérale soit la valeur d'un registre
    if in_lit_operand == 4:
        combo_operand = in_A
    elif in_lit_operand == 5:
        combo_operand = in_B
    elif in_lit_operand == 6:
        combo_operand = in_C
    else:
        combo_operand = in_lit_operand

    return combo_operand
```

```python
def func_opcode_0(in_A, in_combo_operand):
    return in_A //(2**in_combo_operand)
```

```python
def func_opcode_1(in_B, in_lit_operand):
    return in_B ^ in_lit_operand
```

```python
def func_opcode_2(in_combo_operand):
    return in_combo_operand % 8
```

```python
def func_opcode_3(in_A, in_lit_operand, in_index_pointer):
    # Si le registre A vaut 0, on ne fais rien
    if in_A == 0:
        in_index_pointer += 2
        return in_index_pointer
    
    # Sinon on change le pointeur
    new_index_pointer = in_lit_operand

    return new_index_pointer
```

```python
def func_opcode_4(in_B, in_C, in_lit_operand):

    in_B = in_B ^ in_C
    return in_B
```

```python
def func_opcode_5(in_combo_operand):

    return (in_combo_operand % 8)
```

```python
def func_opcode_6(in_A, in_B, in_combo_operand):
    in_B = in_A //(2**in_combo_operand)
    return in_B
```

```python
def func_opcode_7(in_A, in_combo_operand):
    in_C = in_A //(2**in_combo_operand)
    return in_C
```

```python
def func_run_program(in_program, in_A):

    # Initialisation
    index_pointer = 0
    list_out = []
    in_B = 0
    in_C = 0

    # tant que le pointeur reste dans la limite 
    while 0<=index_pointer <= (len(in_program)-2):
        # On récupère l'opcode et l'operand
        opcode = in_program[index_pointer]
        lit_operand = in_program[index_pointer+1]

        # On initialise le combo operand
        combo_operand = func_init_operand(in_lit_operand = lit_operand,
                                          in_A = in_A,
                                          in_B = in_B,
                                          in_C = in_C)

        if opcode == 0:
            in_A = func_opcode_0(in_A = in_A,
                                 in_combo_operand = combo_operand)
            
        elif opcode == 1:
            in_B = func_opcode_1(in_B = in_B,
                                 in_lit_operand = lit_operand)
            
        elif opcode == 2:
            in_B = func_opcode_2(in_combo_operand = combo_operand)
            
        elif opcode == 3:
            index_pointer = func_opcode_3(in_A = in_A,
                                          in_lit_operand = lit_operand,
                                          in_index_pointer = index_pointer)
        elif opcode == 4:
            in_B = func_opcode_4(in_B = in_B,
                                 in_C = in_C,
                                 in_lit_operand = lit_operand)
                        
        elif opcode == 5:
            res_opcode_5 = func_opcode_5(in_combo_operand = combo_operand)
            list_out.append(res_opcode_5)
            if list_out != in_program[:len(list_out)]:
                break
            
        elif opcode == 6:
            in_B = func_opcode_6(in_A = in_A,
                                 in_combo_operand = combo_operand)
            
        elif opcode == 7:
            in_C = func_opcode_7(in_A = in_A,
                                    in_combo_operand = combo_operand)        
        
        if opcode != 3:
            index_pointer += 2


    return list_out    
```
