# Import des librairies

```python
from collections import defaultdict, Counter
```

# Lecture du fichier source

```python
with open('real.txt') as fichier:
    dict_numbers = Counter(map(int,fichier.read().split()))
dict_numbers
```

```python
def func_clignement(in_dict_numbers: dict):
    new_dict = defaultdict(int)
    for number in in_dict_numbers:
        if number == 0:
            new_dict[1] += in_dict_numbers[0]

        else:
            length_number = len(str(number))
            if length_number % 2 == 0:
                new_dict[int(str(number)[:length_number//2])] += in_dict_numbers[number]
                new_dict[int(str(number)[length_number//2:])] += in_dict_numbers[number]
            else:
                new_dict[number * 2024] += in_dict_numbers[number]
    
    return new_dict

```

```python
for i in range(75):
    dict_numbers = func_clignement(dict_numbers)
sum(v for v in dict_numbers.values())
```
