```python
with open('real.txt') as fichier:
    list_stone = [int(e) for e in (fichier.read().split())]
```

```python
def func_apply_rules(in_stone_number:int):

    if in_stone_number == 0 :
        return [1]
    
    str_stone_number = str(in_stone_number)
    nb_chiffres =  len(str_stone_number)
    if nb_chiffres%2 == 0:
        middle_stone_number = nb_chiffres//2
        return [int(str_stone_number[:middle_stone_number]),  int(str_stone_number[middle_stone_number:])]
    
    else:
        return [in_stone_number*2024]
```

```python
def func_clignement(in_list_stone: list[int]):
    result = []
    for stone_number in in_list_stone:
        result.extend(func_apply_rules(stone_number))
    return result

for i in range(32):
    list_stone = func_clignement(list_stone)
print(len(list_stone))
```
