# Fonction qui transforme 12345 ----> 0..111....22222

```python
def func_transform_chain(var_chain : str) -> list:
    # Initialisation
    id = 0
    new_chain = []

    # Pour chaque caractere de la chaine
    for num_char,char in enumerate(var_chain):
        if num_char%2 == 0:
            new_chain.extend([str(id)] * int(char))
            id += 1
        else:
            new_chain.extend(['.'] * int(char))
    return new_chain

```

# Fonction qui transforme 0..111....22222 ----> 022111222......

```python
def func_dot_to_number(var_list:list):
    # On crée une liste qui contient uniquement les nombres (sans les '.') pour itérer ensuite sur le dernier nombre de cette liste par la suite
    only_numbers = [e for e in var_list if e!='.']

    # On récupère la position des '.' dans la liste
    pos_dot = [int(num_char) for num_char,char in enumerate(var_list) if char=='.']
    # Pour chaque position de "." de la liste
    for pos in pos_dot:
        # On remplace le '.' par le dernier nombre de la liste (en ignorant les autres'.' de la fin)
        var_list[pos] = only_numbers[-1]
        # On supprime le nombre que l'on vient de se servir (le dernier nombre de la liste only_number)
        only_numbers.pop()

    # On calcule le nombre de points dans la liste originale
    nb_dot = len(pos_dot)

    # On  enleve tous les nombres que l'on a déplacé pour qu'ils n'apparaissent pas 2 fois dans la liste
    var_list = var_list[:-(nb_dot)]

    # On calcule la somme en parcourant la liste (en s'arretant juste avant )
    return sum((index*int(number)) for index,number in enumerate(var_list))
```
