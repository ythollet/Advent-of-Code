# Fonction qui transforme 12345 ----> 0..111....22222

```python
def func_transform_chain(var_chain : str) -> list[int]:
    """
    Transforme une chaîne comme '12345' en une nouvelle chaîne où, pour chaque caractère,
    si l'indice est pair, on répète un chiffre incrémenté, sinon on insère des points.
    Exemple : '12345' → '0..111....22222'
    ps: les . seront représentés par des -1 cat plus optimisé
    """
    # Initialisation
    id = 0
    new_chain = []

    # Pour chaque caractere de la chaine
    for num_char,char in enumerate(var_chain):
        char = int(char)
        
        # Si impair
        if num_char & 1:
            if char != 0:
                new_chain.extend([None] * char) 
        # Si pair
        else :
            if char != 0:
                new_chain.extend([id] *char)
            id += 1
            
    return new_chain

liste = func_transform_chain(line)
liste
```

```python
relancer = True
while relancer:
    # Continuer le sous traitement
    continuer = True 
    deplacement_effectue = False
    
    # Parcours chaque série de chiffre (fichier) pour les insérer à la place des points
    for pos_number,number in enumerate(reversed(liste)):
        if not continuer:
            break
        pos_number = len(liste) - 1 - pos_number

        # Si on tombe sur un nombre
        if isinstance(number,list) :
            # print(f"\nnombre : {number} | pos : {pos_number}")
            taille_serie = number[1]

            # Parcours tous les points
            for pos_dots,dots in enumerate(liste):

                # On s'arrete si continuer=False ou si les points libres sont situés à droite de notre nombre
                if not continuer or pos_dots>pos_number:
                    break
                if isinstance(dots,str):
                    taille_dots = len(dots)
                    # Si il y a assez de place pour insérer la série de chiffres
                    if taille_serie <= taille_dots:
                        # print(f"\noringinal : {liste}")
                        # On supprime la position de base du nombre
                        liste[pos_number] = -taille_serie
                        # print(f"nombre supprimé : {liste}")

                        # On suppime autant de points que la taille de notre série à insérer
                        # Si il y a pille assez de place
                        if taille_serie == taille_dots:
                            # On vide les anciens points
                            liste[pos_dots] = None
                        # Si il reste de la place après avoir inséré les nombres
                        else:
                            liste[pos_dots] = (taille_dots-taille_serie)*'.'
                            # print(f"points raccourcis: {liste}")
                        # On insère la série de chiffres
                        liste.insert(pos_dots,number)
                        # print(f"nombre insérés : {liste}")
                        liste = [e for e in liste if e!=None]
                        deplacement_effectue = True
                        # print(f"final : {liste}")        
                        continuer = False
                        break
            if deplacement_effectue:
                break
    # Si on est au dernier
    if not relancer:
        relancer = False

# Remplacer les valeurs négatives par des '.'
liste = [-e*'.' if isinstance(e,int) else e for e in liste]
liste
```
