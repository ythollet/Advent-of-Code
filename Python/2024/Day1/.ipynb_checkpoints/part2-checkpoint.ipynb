# Import des librairies

```python
import re
import pandas as pd
import os
import time
```

# Lecture du fichier d'entrée

```python
with open("part2.txt") as text:
    liste=text.readlines()

liste # On obtient une liste de chaines de caractères -------> dans ces chaines se trouvent notamment des nombres
```

# Décodage du fichier d'entrée

```python
# Pour chaque élement de la liste, on récupère les deux nombres ------> le premier est envoyé dans la liste 1 et le 2e dans la liste 2
liste1=[]
liste2=[]
for chaine in liste:
    liste1.append(int(re.findall(r"\d+",chaine)[0]))
    liste2.append(int(re.findall(r"\d+",chaine)[1]))
len(liste1)
```

# Dictionnaire d'effectifs

```python
start = time.perf_counter()
# On crée un dictionnaire qui répertorie pour chaque nombre different contenu dans la liste de gauche, le nombre de fois qu'il apparait dans la liste de droite
ensemble_unique = set(liste1) # Nombre uniques contenus dans la liste de gauche
dico_effectifs = {cle:liste2.count(cle) for cle in ensemble_unique}
dico_effectifs

sum([(key*values) for key,values in dico_effectifs.items()])
end = time.perf_counter()
print(end-start)
```

# Calcul du score de similarité

```python
sum([(key*values) for key,values in dico_effectifs.items()])
```

# Calcul du score de similarité (solution)

```python
start = time.perf_counter()

total=0
for nb in liste1:
    total += nb*liste2.count(nb)
total

end = time.perf_counter()
print(end-start)
```
