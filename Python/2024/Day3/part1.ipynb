# Import des bibliothèques

```python
import re
from functools import reduce
```

# Lecture du fichier

```python
with open ("real.txt") as file:
    fichier = file.read()
fichier
```

# Mise en forme et taitement des données

```python
[reduce(lambda x,y: int(x)*int(y),re.findall(r'\d+',e)) for e in re.findall(r'mul\(\d+\,\d+\)',fichier)]
```

```python
sum(int(nb1) * int(nb2) for nb1, nb2 in re.findall(r'mul\((\d+),(\d+)\)', fichier))
```
