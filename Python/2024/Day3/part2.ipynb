# Import des bibliothèques

```python
import re
from functools import reduce
```

# Lecture du fichier

```python
with open ("real2.txt") as file:
    fichier = file.read()
fichier
```

# On recupère les mul(...,...)

```python
list_mult = [[int(nb1)*int(nb2),mul.start()]
             for mul in re.finditer(r'mul\((\d+),(\d+)\)', fichier) # récupère les mul(512,255)...
             for nb1,nb2 in [mul.groups()] # récupère les nombres contenus dans mul(622,262)
            ]
# nb1 * nb2 | start position
list_mult
```

# On récupère les do et don't

```python
do_dont = ([["do",do.start()] for do in re.finditer(r'do\(\)',fichier)]+
          [["dont",dont.start()] for dont in re.finditer(r"don\'t\(\)",fichier)])

# On trie par ordre de position du fichier source
do_dont = sorted(do_dont,key=lambda x:x[1])

do_dont
```

```python
# On optimise la liste ---> si on a un do()...do()...dont() on simplifie en do()...dont()
# On recupere les indices à supprimer
prev_statut = do_dont[0][0]
pos_supp=[]


for statut,pos in do_dont[1:]:
    if prev_statut == statut:
        pos_supp.append(pos)
    prev_statut = statut
pos_supp

# On supprimes les élements redondants
clean_list = [liste[1] for liste in do_dont if liste[1] not in pos_supp]


# Output : dont:318
#          do:869.... 
#          do...
#          dont....
```

On a maintenant la liste des do/dont sans doublons successifs

```python
# On regroupe le premier do avec le premier don't ect...
group_clean_list = list(zip(clean_list[::2],clean_list[1::2]))
group_clean_list
```

On doit doit reparcourir le fichier source en supprimmant les multiplication entre la position 318 et 869 puis entre 1528 et 1751...

# On reparcourt le fichier

```python
# On parcours la liste de multiplications
list_supp_pos_mult = []
for multiplication,pos_mult in list_mult:
    for intervalle in group_clean_list:
        # Si la position est comprise dans l'intervalle
        if intervalle[0] < pos_mult < intervalle[1]:
            list_supp_pos_mult.append(pos_mult)

sum(mult for mult,pos_mult in list_mult if pos_mult not in list_supp_pos_mult)
```
