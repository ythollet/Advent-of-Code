# Lecture du fichier source

```python
with open ('real.txt') as fichier:
    grille = fichier.read().split("\n")
    
grille
```

```python
list_dep = [(num_line, num_col)
        for num_line,line in enumerate(grille)
        for num_col,char in enumerate(line)
        if char == '0'
        ]

list_dep
```

# Fonction qui regarde autour

```python
def func_check_around(number : int,
                      pos_number : tuple,
                      grille : list
                     ):

    # Récupère les dimensions de la grille
    dim_grille = len(grille)

    # Conversion du nombre en str pour comparer des str avec des str
    next_number_str = str(number+1)
    # print('current number',number)

    res = []

    # Regarder à gauche (-1,0), droite (1,0), haut (0,-1), bas (0,1)
    for decallage in [(-1,0),(1,0),(0,-1),(0,1)]:
        v = pos_number[0] + decallage[0]
        h = pos_number[1] + decallage[1]
        if 0 <= v <= (dim_grille-1) and 0 <= h <= (dim_grille-1):
            # print(f'v:{v} h:{h}')            
            if grille[v][h] == next_number_str:
                # print('yes',grille[v][h],next_number_str)
                
                res.append((v,h))
    
    return res

```

# Fonction récursive qui explore tous les chemins (arrivants à 9) possibles pour un 0 donné

```python
def func_forward_and_check(current_number : int,
                           current_pos_number : tuple,
                           grille : list,
                           path: list = [] ,
                           visited_nine = None
                           ):
    path.append(current_pos_number)
    all_path=[]

    if visited_nine is None:
        visited_nine = []

    # Un chemin complet a été trouvé, fin d'une branche de la récursion
    if current_number == 9 and current_pos_number not in visited_nine:
        visited_nine.append(current_pos_number)
        # print("visited nine",visited_nine)
        return [path[:]]    

    # On regarde si il y a des possibilités autour
    list_res_around = func_check_around(current_number, current_pos_number, grille)

    # Pour chacune de ces possiblités
    for res_around in list_res_around:
        # On appelle la fonction sur cette possibilité, puis on stocke le chemin correspondant dans all_path
        all_path.extend(func_forward_and_check(current_number+1, res_around, grille, path[:], visited_nine))
            
    return all_path

```

# On boucle sur tous les points de départs

```python
score = 0
for point_dep in list_dep:
    print(f'point_dep : {point_dep}')
    differents_chemins = func_forward_and_check(0,point_dep,grille)
    print(f'différents chemins :')
    print(differents_chemins)

    score += len(differents_chemins)

score
```
