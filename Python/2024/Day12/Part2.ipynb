```python
def func_proche(pos1:tuple,
                pos2:tuple,
                ):
    
    if abs(pos1[0]-pos2[0]) + abs(pos1[1]-pos2[1]) == 1:
        return True
    
    return False
```

```python
def func_integrate_to_group(in_group:list,
                            in_to_integrate:list):
    
    # Pour chacun des emplacements de la lettre
    for pos in in_to_integrate:
        
        # Pour chaque emplacements validé de la lettre 
        for pos_validate in in_group:
            # Est ce que elle est assez proche pour etre intégrée au groupe
            if func_proche(pos,pos_validate):
                in_group.append(pos)
                in_to_integrate.remove(pos)
                return in_group, in_to_integrate

    return in_group, in_to_integrate
    

```

```python
# Cette fonction trouve un groupe et revoie ce dernier + le reste a detecter
def func_find_a_group(in_list_pos_letter:list):

    group = [in_list_pos_letter[0]]
    to_integrate = in_list_pos_letter[1:]
    continuer = True
    previous_len = 0

    while continuer:
        group,to_integrate = func_integrate_to_group(group,to_integrate)
        # Si aucune modif n'as été faite
        if to_integrate==[] or previous_len == len(group):
            continuer = False
        previous_len = len(group)

    return group, to_integrate
```

```python
def func_around(in_pos:tuple,
                in_taille_grille:int):
    return [pos for pos in [(in_pos[0]+1,in_pos[1]), (in_pos[0]-1,in_pos[1]), (in_pos[0],in_pos[1]+1), (in_pos[0],in_pos[1]-1)]]
```

```python
def func_all_groups_for_a_letter(in_list_pos_letter:list):

    to_integrate=in_list_pos_letter
    group=[]
    groups=[]
    while to_integrate!=[]:
        group, to_integrate=func_find_a_group(to_integrate)
        # On ajoute le groupe récupéré à la liste des groupes de la lettre
        groups.append(group)


    return groups
```

```python
def func_dim_matrix(in_data:list[tuple[int,int]]):
    """
    Calcule les dimensions d'une matrice

    """
    min_row = max_row = in_data[0][0]
    min_col = max_col = in_data[0][1]

    for row, col in in_data[1:]:
        min_row = min(min_row, row)
        max_row = max(max_row, row)
        min_col = min(min_col, col)
        max_col = max(max_col, col)

    return min_row, max_row, min_col, max_col
```

```python
def func_verif_complete_matrix(in_data:list[tuple[int,int]]):
    """
    Vérifie si une une parcelle est rectangulaire (ou carrée)
    """

    min_row, max_row, min_col, max_col = func_dim_matrix(in_data = in_data)

    nrow = 1 + max_row - min_row
    ncol = 1 + max_col - min_col

    nb_elements_attendus =  nrow * ncol
    if len(in_data) < nb_elements_attendus:
        return False
    
    return True
```

```python
# prends en parametre les positions des lettres d'une parcelle  et renvoie le prix correspondant
def func_calcul_price(in_pos_letter_parcelle:list[tuple[int,int]],
                      in_taille_grille:int):

    # print(f"fim_groups_stangers{data[in_pos_letter_parcelle[0][0]][in_pos_letter_parcelle[0][1]]}")
    perimetre = func_calcul_perimetre(in_pos_letter_parcelle, in_taille_grille)
    aire = len(in_pos_letter_parcelle)
    prix = aire * perimetre
    
    print(f'Prix : {aire} * {perimetre} = {prix}')
    print()
    return prix
```
