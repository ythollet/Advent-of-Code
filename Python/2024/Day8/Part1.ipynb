# Import des librairies

```python
from collections import defaultdict
from itertools import combinations
```

# Lecture du fichier

```python
with open('real.txt') as file:
    fichier = file.read().split('\n')
fichier
```

# Nettoyage

```python
dico = defaultdict(set)

for num_line,line in enumerate(fichier):
    for index,char in enumerate(line):
        if char !='.':
            dico[char].add((num_line, index))

dico
```

# Fonction qui vérifie si une position est dans la grille

```python
def func_in_grille(var_row_a_verif : int,
                   var_col_a_verif : int,
                   var_nb_row_grille : int,
                   var_nb_col_grille : int)->bool:
    print(var_nb_row_grille,var_nb_col_grille)
    if (0 <= var_row_a_verif < var_nb_row_grille) and (0 <= var_col_a_verif < var_nb_col_grille):
        return True
    return False
```

```python
nb_col_grille = len(fichier)
nb_row_grille = len(fichier[0])
antinodes=set()
# Pour chaque typle d'antenne
for type_antenne, liste_pos_antenne in dico.items():
    # Pour chaque combinaison de deux positions
    print(type_antenne)
    for combo in combinations(liste_pos_antenne,2):
        print("combo",combo)
        pos1,pos2 = combo
        # On calcule la différence Verticale et Horizontale entre les deux positions    
        diff = (pos1[0]-pos2[0], pos1[1]-pos2[1])

        # 1 - On ajoute la difference à la premiere position
        new_obstacle1 = ((pos1[0]+diff[0]),(pos1[1]+diff[1]))
        if func_in_grille(var_row_a_verif = new_obstacle1[0],
                          var_col_a_verif = new_obstacle1[1],
                          var_nb_row_grille = nb_row_grille,
                          var_nb_col_grille = nb_col_grille
                          ):
            print("nouvel obstacle",new_obstacle1)
            antinodes.add(((pos1[0]+diff[0]),(pos1[1]+diff[1])))
        else:
            print(new_obstacle1,'not in grille')

        # 2 - On retire la difference a la deuxieme position
        new_obstacle2 = ((pos2[0]-diff[0]),(pos2[1]-diff[1]))    
        if func_in_grille(var_row_a_verif = new_obstacle2[0],
                          var_col_a_verif = new_obstacle2[1],
                          var_nb_row_grille = nb_row_grille,
                          var_nb_col_grille = nb_col_grille
                         ):
            print("nouvel obstacle 2",new_obstacle2)
            antinodes.add(((pos2[0]-diff[0]),(pos2[1]-diff[1])))
            print('\n')
        else:
            print(new_obstacle1,'not in grille')
len(antinodes)
```
