# Fonction qui vérifie si une position est dans la grille

```python
def func_in_grille(var_row_a_verif : int,
                   var_col_a_verif : int,
                   var_nb_row_grille : int,
                   var_nb_col_grille : int)->bool:
    return (0 <= var_row_a_verif < var_nb_row_grille) and (0 <= var_col_a_verif < var_nb_col_grille)
```

# Fonction Traitement Horizontal

```python
def func_traitement_horizontal(var_pos1:tuple,
                               var_pos2:tuple,
                               var_y_start:int,
                               var_y_step:int,
                               var_y_stop:int)->set:
    res = set()    
    for y in range(var_y_start, var_y_stop, var_y_step):
        obstacle = (0,y)
        res.add(obstacle)
    return res
```

# Fonction Traitement Vertical

```python
def func_traitement_vertical(var_pos1:tuple,
                             var_pos2:tuple,
                             var_x_start:int,
                             var_x_step:int,
                             var_x_stop:int)->set:
    res = set()    
    for x in range(var_x_start, var_x_stop, var_x_step):
        obstacle = (x,0)
        res.add(obstacle)
    return res
```

# Fonction Traitement Diagonal

```python
def func_traitement_diagonal(var_pos1:tuple,
                             var_pos2:tuple,
                             var_x_start:int,
                             var_x_step:int,
                             var_x_stop:int,
                             var_y_start:int,
                             var_y_step:int,
                             var_y_stop:int)->set:
    res = set()                         
    for x,y in zip(range(var_x_start, var_x_stop, var_x_step),range(var_y_start, var_y_stop, var_y_step)):
        obstacle = (x,y)
        res.add(obstacle)
    return res
```

# Fonction qui cherche les obstacle pour 2 positions

```python
def func_find_obstacles(var_pos1 : tuple,
                        var_pos2 : tuple,
                        var_dim_grille,
                        ) -> set:
    
    diff = (var_pos2[0]-var_pos1[0], var_pos2[1]-var_pos1[1])


    # 1 - Sur la diagonale à traiter on va dabord s'occuper du coté de la première position

    # Start 
    x_start, y_start = var_pos1

    # Step
    x_step = -diff[0]
    y_step = -diff[1]


    # Stop
    if x_step < 0:
        x_stop = -1
    else : 
        x_stop = var_dim_grille

    if y_step < 0:
        y_stop = -1
    else : 
        y_stop = var_dim_grille


    res=set()

    # Si les deux positions sont alignées à l'horizontale
    if x_step == 0:
        res|=func_traitement_horizontal(var_pos1 = var_pos1,
                                        var_pos2 = var_pos2,
                                        var_y_start = y_start,
                                        var_y_step = y_step,
                                        var_y_stop = y_stop)

    # Si les deux positions sont alignées à la verticale
    elif y_step == 0:
        res|=func_traitement_vertical(var_pos1 = var_pos1,
                                      var_pos2 = var_pos2,
                                      var_x_start = x_start,
                                      var_x_step = x_step,
                                      var_x_stop = x_stop)

    # Si les deux positions sont alignées en diagonale
    elif x_step!=0 and y_step!=0:
        res|=func_traitement_diagonal(var_pos1 = var_pos1,
                                 var_pos2 = var_pos2,
                                 var_x_start = x_start,
                                 var_x_step = x_step,
                                 var_x_stop = x_stop,
                                 var_y_start = y_start,
                                 var_y_step = y_step,
                                 var_y_stop = y_stop)
    


    # 2- Après avoir fait le premier sens (celui qui va du coté de la premiere position), on fait l'autre sens
    
    # Start 
    x_start = var_pos2[0] 
    y_start = var_pos2[1] 

    # Step
    x_step = diff[0]
    y_step = diff[1]
    


    # Stop
    if x_step < 0:
        x_stop = -1
    else : 
        x_stop = var_dim_grille

    if y_step < 0:
        y_stop = -1
    else : 
        y_stop = var_dim_grille


    # Si les deux positions sont alignées à l'horizontale
    if x_step == 0:
        res|=func_traitement_horizontal(var_pos1 = var_pos1,
                                        var_pos2 = var_pos2,
                                        var_y_start = y_start,
                                        var_y_step = y_step,
                                        var_y_stop = y_stop)

    # Si les deux positions sont alignées à la verticale
    elif y_step == 0:
        res|=func_traitement_vertical(var_pos1 = var_pos1,
                                      var_pos2 = var_pos2,
                                      var_x_start = x_start,
                                      var_x_step = x_step,
                                      var_x_stop = x_stop)

    # Si les deux positions sont alignées en diagonale
    elif x_step!=0 and y_step!=0:
        res|=func_traitement_diagonal(var_pos1 = var_pos1,
                                 var_pos2 = var_pos2,
                                 var_x_start = x_start,
                                 var_x_step = x_step,
                                 var_x_stop = x_stop,
                                 var_y_start = y_start,
                                 var_y_step = y_step,
                                 var_y_stop = y_stop)


    return res
```

# Traitement

```python
dim_grille = len(fichier)
antinodes=set()
# Pour chaque typle d'antenne
for type_antenne, liste_pos_antenne in dico.items():
    # Pour chaque combinaison de deux positions
    for combo in combinations(liste_pos_antenne,2):
        pos1,pos2 = combo
        # On récupères les antinodes correspondants à ces deux positions
        res_func_antinodes = func_find_obstacles(var_pos1 = pos1,
                                                 var_pos2 = pos2,
                                                 var_dim_grille =  dim_grille
                                                 )
        # On ajoute uniquement les nouveaux antinodes
        if res_func_antinodes:
            antinodes |= res_func_antinodes

len(antinodes)
```

```python
with open("real.txt") as fin:
    grid = fin.read().strip().split("\n")

n = len(grid)

def in_bounds(x, y):
    return 0 <= x < n and 0 <= y < n

def get_antinodes(a, b):
    ax, ay = a
    bx, by = b
    
    dx, dy = bx - ax, by - ay

    i = 0
    while True:
        if in_bounds(ax - dx * i, ay - dy * i):
            yield (ax - dx * i, ay - dy * i)
        else:
            break
        i += 1
    
    i = 0
    while True:
        if in_bounds(bx + dx * i, by + dy * i):
            yield (bx + dx * i, by + dy * i)
        else:
            break
        i += 1


antinodes = set()

all_locs = defaultdict(list)
for i in range(n):
    for j in range(n):
        if grid[i][j] != ".":
            all_locs[grid[i][j]].append((i, j))


for freq in all_locs:
    locs = all_locs[freq]
    for a, b in combinations(locs, r=2):
        for antinode in get_antinodes(a, b):
            antinodes.add(antinode)


for i in range(n):
    for j in range(n):
        if (i, j) in antinodes:
            print("#", end="")
        else:
            print(grid[i][j], end="")
    print()


print(len(antinodes))
```

```python
def nombres_paires(n):
    for i in range(n):
        if i % 2 == 0:
            yield i

for nombre in nombres_paires(10):
    print(nombre) 

```
