# Import des librairies

```python
import numpy as np
from heapq import heappush, heappop
```

# Lecture du fichier source

```python
with open('real.txt') as fichier:
    input = [tuple(map(int,coord.split(','))) 
            for coord 
            in fichier.read().split('\n')]
```

# Préparation des données

```python
limit_grid = 70
num_firsts_bytes = 1024
obstacles = input[:num_firsts_bytes]


# Créer la grille avec numpy
grille = np.zeros((limit_grid+1, limit_grid+1), dtype=int)

# Ajout des obstacles
for x_obs,y_obs in obstacles[:num_firsts_bytes]:
    grille[y_obs,x_obs] = 1

grille
```

# Traitement

```python
def func_in_grid(in_pos,
                 in_dim_grid):
    if (0 <= in_pos[0] <= in_dim_grid) and (0 <= in_pos[1] <= in_dim_grid):
        return True
    return False
```

```python
def func_pos_around(in_current_pos,
                    in_obstacles):
    
    res = []
    list_delta = [(-1,0),(0,1),(1,0),(0,-1)]
    for delta in list_delta:
        new_pos = (in_current_pos[0] + delta[0], in_current_pos[1] + delta[1])
        if new_pos not in in_obstacles:
            if func_in_grid(new_pos, limit_grid):
                res.append(new_pos)
    return res
```

```python
file = [(0,0,0)]
final_pos = (limit_grid,limit_grid)
visited = set()
distances = {(y,x) : float('inf') for x in range(limit_grid+1) for y in range(limit_grid+1)}
distances[(0,0)] = 0

while file:
    # On récupère la prochaine position dans la file
    num_row,num_col,current_dist = heappop(file)
    pos = (num_row,num_col)

    # Si la position a deja été visitée
    if pos in visited:
        continue
    
    # On marque la position comme visitée
    visited.add(pos)

    # Si on a atteint la position cible
    if pos == final_pos:
        print(current_dist)
    
    # On récupère les positions autour
    list_pos_around = func_pos_around(pos, obstacles)
    for pos_around in list_pos_around:
        new_distance = current_dist + 1
        if new_distance < distances[pos_around]:
            distances[pos_around] = new_distance 
            heappush(file,(pos_around[0], pos_around[1], new_distance))
    
```

```python
from heapq import heappush, heappop

with open("real.txt") as fin:
    lines = fin.read().strip().split("\n")
    coords = set([tuple(map(int, line.split(","))) for line in lines][:1024])

dd = [[1, 0], [0, 1], [-1, 0], [0, -1]]

N = 70

# Heuristic for A*
def h(i, j):
    return abs(N - i) + abs(N - j)

def in_grid(i, j):
    return 0 <= i <= N and 0 <= j <= N and (i, j) not in coords

q = [(h(0, 0), 0, 0)]
cost = {}
while len(q) > 0:
    c, i, j = heappop(q)
    if (i, j) in cost:
        continue
    cost[(i, j)] = c - h(i, j)

    if (i, j) == (N, N):
        print(cost[(i, j)])
        break

    for di, dj in dd:
        ii, jj = i + di, j + dj
        if in_grid(ii, jj):
            heappush(q, (cost[(i, j)] + 1 + h(ii, jj), ii, jj))
```
