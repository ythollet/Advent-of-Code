# Lecture du fichier

```python
with open ("real.txt") as fichier:
    file = fichier.readlines()
file 
```

# Nettoyage du fichier source

```python
# On enleve les /n
file=[e.strip() for e in file]
file.remove('')
file

# On sépare le fichier en 2
liste1, liste2 = [],[]
for e in file:
    if "|" in e:
        numbers = e.split("|")
        numbers = [int(e) for e in numbers]
        liste1.append(numbers)
    else:
        numbers = e.split(",")
        numbers = [int(e) for e in numbers]
        liste2.append(numbers)

liste1,liste2
```

# Fonction qui vérifie si une liste est trié

```python
# Pour chaque liste contenue dans liste2

def func_verif_order(liste2_nb):
    valide = True # on part du principe qu'initialement, la liste est bien classée
    # Pour chaque nombre de cette liste
    for pos_nb,nb in enumerate(liste2_nb):
        # Vérifie si le nombre étudié est bien classé
        # On recupere les nombres placés devant le nombre que l'on étudie
        nombres_a_verifier = [liste[0] for liste in liste1 if nb == liste[1]]
        # print(f"nombres a verifier : {nombres_a_verifier}")
        # Pour chaucune de ces sous listes où le nombre en question est placé en deuxième position, on vérifié si il est contenu dans la liste du nombre que l'on étudie 
        for nb_a_verifier in nombres_a_verifier:
            # Si le nombre à vérifier est contenu dans la liste du nombre de base
            if nb_a_verifier in liste2_nb:
                # On recupere sa position dans la liste2
                pos_nb_a_verifier = liste2_nb.index(nb_a_verifier)
                # Si il est placé avant le nombre que l'on étudie, la liste est fausse
                if pos_nb < pos_nb_a_verifier:
                    valide=False
                    # print(f" nb : {nb} en position {pos_nb}\n nombre a vérifier : {nb_a_verifier} en position {pos_nb_a_verifier}")
    return valide
            
```
