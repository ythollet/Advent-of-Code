# Import des librairies

```python
import numpy as np
import re
```

# Lecture du fichier

```python
with open ("real.txt") as fichier:
    file = np.array(fichier.readlines())
file 
```

# Nettoyage du fichier source

```python
file = np.array([list(line.replace("\n",""))
        for line 
        in file
    ])
file
```

# Traitement horizontal

```python

def func_horizontal(array):
    array = ["".join(line) for line in array]
    return sum(line.count("XMAS") for line in array) + sum(line.count("SAMX") for line in array)

func_horizontal(file)
```

# Traitement vertical

```python
# Au lieu de faire un traitement vertical, on traspose la matrice et on effectue un traitement horizontal
def func_verticale(array):
    array = array.transpose()
    return func_horizontal(array)

func_verticale(file)
```

# Traitement diagonales descendentes

```python
# Parcourt toutes les diagonales
def func_diag_descendentes(array):
    diagonales = []
    print(array)
    for i in range (1-len(array),len(array)):
        diagonales.append(array.diagonal(i))
    return func_horizontal(diagonales)

func_diag_descendentes(file)
```

# Traitement diagonales montantes


```python
def func_diag_montantes(array):
    array = np.rot90(array)
    print(array)
    return func_diag_descendentes(array)
func_diag_montantes(file)
```

# Traitement regroupé

```python
func_horizontal(file) + func_verticale(file) + func_diag_descendentes(file) + func_diag_montantes(file)
```
