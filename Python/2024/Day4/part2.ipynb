# Import des librairies

```python
import numpy as np
import re
```

# Lecture du fichier

```python
with open ("real2.txt") as fichier:
    file = np.array(fichier.readlines())
file 
```

# Nettoyage du fichier source

```python
file = np.array([list(line.replace("\n",""))
        for line 
        in file
    ])
file
```

# Positions des A

```python
# On recupere la position des A en ommettant ceux qui sont au bord du tableau
pos_a = [(l,i) for l in range(1,len(file)-1) for i in range(1,len(file[l])-1) if file[l][i]=="A"]
```

# On reparcours le fichier en bouclant sur les pos de A

```python
# On vérifie si on a bien une croix de MAS ou SAM avec le A au milieu

compteur = 0 # compte le nombre de A corrects
for l,c in pos_a:
    temp=[] 
    # On recupère les lettres autour du A
    temp=[file[l-1][c-1], file[l+1][c+1], file[l-1][c+1], file[l+1][c-1]]
    # On regarde si on a 2 M et 2 S
    if temp.count("M") == 2 and temp.count("S") == 2:
        # On regarde si ces 2M et 2S ne se suivent pas en diagonale (ex: SAS ou MAM)
        if temp[0] != temp[1]:
            compteur+=1
compteur
    

```
